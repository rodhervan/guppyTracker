```python
# import necessary libraries 
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Perspective images/t1.jpg')
frame = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
frame.shape
orig = frame.copy()

tl = (480,420); tr = (910,405)
bl = (432,629); br = (950,610)

cv2.circle(frame,tl,10,(255,0,0),-1);cv2.circle(frame,tr,10,(255,0,0),-1)
cv2.circle(frame,bl,10,(255,0,0),-1);cv2.circle(frame,br,10,(255,0,0),-1)

s = frame.shape
pts1 = np.float32([tl, bl, tr, br])
pts2 = np.float32([[0,0],[0,s[0]],[s[1],0],[s[1],s[0]]])

matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(orig, matrix, (s[1], s[0]))

plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.imshow(frame) 
plt.subplot(1,2,2)
plt.imshow(result)
```
